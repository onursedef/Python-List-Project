## Python Projesi

#### Sorular:
* Bir listeyi düzleştiren (flatten) fonksiyon yazın. Elemanları birden çok katmanlı listelerden ([[3],2] gibi) oluşabileceği gibi, non-scalar verilerden de oluşabilir. Örnek olarak:
    * input: `[[1,'a',['cat'],2],[[[3]],'dog'],4,5]`
    * output: `[1,'a','cat',2,3,'dog',4,5]`
* Verilen listenin içindeki elemanları tersine döndüren bir fonksiyon yazın. Eğer listenin içindeki elemanlar da liste içeriyorsa onların elemanlarını da tersine döndürün. Örnek olarak:
    * input: `[[1, 2], [3, 4], [5, 6, 7]]`
    * output: `[[[7, 6, 5], [4, 3], [2, 1]]`

#### İlk Soru
Öncelikle denemek için bir liste oluşturalım.

```python
arr = [[1,'a',['cat'],2],[[[3]],'dog'],4,5]
arr
```

Listemiz görünüyor. Listemizi düzeltecek fonksiyonu yazmaya geçebiliriz.

```python
from collections.abc import Iterable

def beaufity_arr(arr):
    for i in arr:
        if type(i) is list:
            for e in i:
                if isinstance(e, Iterable) and  not isinstance(e, str):
                    beaufity_arr(e)
                else:
                    yield e
        else:
            yield i
```

Burada Python modülü olan `collections.abc`'den `Iterable`'ı çektik, çünkü değerin yinelebilir olup olmadığını kontrol etmemiz gerekiyor.
Fonksiyonun nasıl işlediğine gelirsek:
1. Öncelikle bir fonksiyonumuzda iki adet değer, girdi vermek zorundayız. Bunlardan biri dağınık listemiz diğeri düzeltimiş yeni listemiz olacak.
2. Ardında her bir değeri tek tek kontrol etmemiz gerekiyor. 
3. Eğer değer tipi list, tuple veya dictionary ise bu değerin içindeki değerleri tek tek tekrar kontrol etmemiz gerekiyor. Eğer bu değerler değilse, değerleri ekrana yazdırıyoruz.
4. Eğer bu listenin içindeki değerler yinelebilir ve string değilse 2'inci adıma geri dönüyoruz.
5. Eğer yenilenemez veya string ise değerleri ekrana yazdırıyoruz.

```python
list(beaufity_arr(arr))
```

#### İkinci soru
Aslında ilk soruya benzyor fakat bunda liste düzeltmek yerine listeyi ve elemanlarını ters çevirmemiz gerekiyor.
Denemek için bir liste tanımlayalım.

```python
normal_arr = [[1, 2], [3, 4], [5, 6, 7]]
normal_arr
```

Şimdi ise bu listeyi ve bu listenin içindeki listeyi ters çevirecek fonksiyonu yazalım.

```python
def reverse_arr(arr):
    for i in arr:
        if type(i) is list:
            reverse_arr(i)
    arr.reverse()
    return arr
```

Bu fonksiyonda önceki fonksiyon gibi işliyor.
1. Bir liste alıyoruz ve onu elemanlarına ayırıyoruz.
2. Listeyi tersine çeviriyoruz.

3. Elemanlardan liste olanları ayırıp tekrar 1. adıma dönüyoruz.

#### Sonuç:

```python
reverse_arr(normal_arr)
```
